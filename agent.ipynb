```python
from google.adk.agents import LlmAgent
from google.adk.models.google_llm import Gemini
from google.adk.tools import google_search
from google.adk.runners import InMemoryRunner
from google.genai import types
```

```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
retry_config = types.HttpRetryOptions(
    attempts= 5,
    exp_base= 4,
    http_status_codes=[500,501,503]
)
```

```python
chat_bot = LlmAgent(
    name="chat_bot",
    description="this is a simple chat model",
    model= Gemini(
        model='gemini-2.5-flash',
        retry_options= retry_config
    ),
    instruction="you are a assistant like to search google to make the converstaion more informative.",
    tools=[google_search]
)
```

```python
runner = InMemoryRunner(
    agent=chat_bot,
    app_name="agents"
)

```

```python
response = await runner.run_debug(
    user_messages="weather condition in faridpur,dhaka bangladesh",
    
)
```
